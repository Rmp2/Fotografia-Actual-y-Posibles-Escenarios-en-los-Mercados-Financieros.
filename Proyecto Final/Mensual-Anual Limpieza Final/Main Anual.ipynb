```python
spx = pd.read_csv('SPX.csv')
spx.head()
```

```python
spx = spx[(spx['Date'] > '1989-12-31')]
```

```python
gold_date=pd.to_datetime(spx['Date'], format='%Y/%m/%d')
gold_date=pd.DataFrame(gold_date)
```

```python
spx2=spx.drop(columns='Date', axis=1)
```
