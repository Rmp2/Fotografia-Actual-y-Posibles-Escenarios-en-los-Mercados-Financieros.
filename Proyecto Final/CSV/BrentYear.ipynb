```python
import pandas as pd
```

```python
bren=pd.read_csv('brentY.csv')
bren
```

```python
bren['Date'][0]
```

```python
br=[i.split('-') for i in bren['Date']]
br
```

```python
bre=[i for a in br for i in a]
año=bre[0::3]

di={'Año':año}
```

```python
bren1=pd.DataFrame(di)

bren2=pd.concat([bren1,bren], axis=1)

bren2
```

```python
brent=bren2.drop('Date', axis=1)
brent
```
