```python
import pandas as pd
```

```python
spx=pd.read_csv('HistoricalData_SPX.csv')
spx
```

```python
spx1=spx[['Cerrar/último']]
```

```python
spx2=pd.to_datetime(spx['Fecha'])
spx2=pd.DataFrame(spx2)
spx2
```

```python
sp=pd.concat([spx1,spx2], axis=1)
sp
```

```python
#sp.rename(columns={'Cerrar/último': 'Close', 'Fecha': 'Date'}, inplace=True) Para Fibonacci
```

```python
#sp.to_csv('SPX.csv', index=False) 
```

```python
sp['Month'] = pd.DatetimeIndex(spx2['Fecha']).month
sp['Year'] = pd.DatetimeIndex(spx2['Fecha']).year
sp
```

```python
sp2 = sp[(sp['Fecha'] >= '12/01/2020')]
sp2
```

```python
s=sp2.drop_duplicates(subset=['Month','Year'],keep='last')
s=s.drop(['Fecha'], axis=1)
s
```

```python
s=s.rename({'Cerrar/último': 'SPX_value'}, axis=1)
s
```
