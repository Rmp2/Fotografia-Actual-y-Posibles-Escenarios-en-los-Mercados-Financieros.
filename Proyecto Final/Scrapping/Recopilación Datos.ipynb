```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import regex as re
```

```python
#Tabla de Precios de la electricidad para consumidores domésticos
url='https://watiofy.com/info/blog/comparativa-precio-luz-europa/'
browser=webdriver.Chrome()
browser.get(url)
```

```python
eve=browser.find_elements(By.TAG_NAME,'tbody')
bank=[i.text for i in eve]
bank
```

```python
bank2=[i for a in bank for i in a.split('\n')]
bank2
```

```python
regex=r'\w.+'
y=[]
for i in bank2:
    y.append(re.findall(regex,i))
y
```

```python
bank=[i for a in y for i in a]
#bank.sort(reverse=True) #Ordenar los valores
bank.remove(bank[0])
bank
bank.remove(bank[0])
bank
bank.remove(bank[15])
bank
bank.remove(bank[14])
bank
bank.remove(bank[19])
bank
```

```python
ban=[i for a in bank for i in a.split(" ")]
ban
```

```python
Paises=ban[0::4]
Tri1=ban[1::4]
Tri2=ban[2::4]
Variacion=ban[3::4]

Tri1= [re.sub(',', '.', i) for i in Tri1]
Tri1=[float(i) for i in Tri1]

Tri2= [re.sub(',', '.', i) for i in Tri2]
Tri2=[float(i) for i in Tri2]
```

```python
dic={'1ºSemestre 2021':Tri1,
    '2ºSemestre 2021':Tri2,
    'Variación':Variacion}

coDomestico=pd.DataFrame(dic, index=Paises)
coDomestico
```

### España - Consumo de electricidad

```python
energia=[i for a in ener for i in a.split(' ')]
energia
```

```python
Año=energia[::4]
Gene=energia[1::4]
Con=energia[2::4]
cap=energia[3::4]

Año=Año[0:21]
Gene=Gene[0:21]
Con=Con[0:21]
cap=cap[0:21]
```

### inflación histórica del IPC España (anual)

```python
url='https://www.inflation.eu/es/tasas-de-inflacion/espana/inflacion-historica/ipc-inflacion-espana.aspx'
brow=webdriver.Chrome()
brow.get(url)
```

```python
des=brow.find_elements(By.CLASS_NAME,'tabledata1')
des2=brow.find_elements(By.CLASS_NAME,'tabledata2')
des=[i.text for i in des[0:5]]
des2=[i.text for i in des2[0:5]]
```

```python
combdesp= [i.strip().split('   ') for a in zip(des, des2)for i in a]
combdesp
```

```python
combdesp2=[i for a in combdesp for i in a]
combdesp2.sort(reverse=True)
combdesp2
```

```python
l=[i for a in combdesp2 for i in a.split(' ')]
pais=l[1::5]
ano=l[2::5]
cant=l[3::5]
```

```python
a= [re.sub(',', '.', i) for i in cant]
b=[float(i) for i in a]

dicto={
      'InflaciónEspaña':b}
```

```python
InflacionEsp=pd.DataFrame(dicto, index=ano)
InflacionEsp
```

```python
#InflacionEsp.to_csv('InflacionEsp.csv', index=False)
```

```python
e=[i for a in combeu2 for i in a.split(' ')]
Pais=e[1::5]
Ano=e[2::5]
Cant=e[3::5]
```
