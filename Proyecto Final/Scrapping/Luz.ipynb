```python
import pandas as pd
import numpy as np
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import regex as re
```

```python
#evolución de la tarifa regulada en los meses anteriores y posteriores a la entrada en vigor de las nuevas tarifas (en junio de 2021)
url='https://www.ocu.org/vivienda-y-energia/gas-luz/informe/precio-luz'
brow=webdriver.Chrome()
brow.get(url)
```

```python
eve=brow.find_elements(By.TAG_NAME,'table')
tar=[i.text for i in eve]
tar
```

```python
tarjet=[i for a in tar for i in a.strip().split('\n')]
tarjet.remove(tarjet[1])
tarjet
```

```python
tarjet.remove(tarjet[1])
tarjet1=[i for a in tarjet for i in a.strip().split(' ')]
```

```python
tarjet1=[i for a in tarjet1 for i in a.strip().split(' ')]
tarjet1
```

```python
for i in tarjet:
    if i == '':
        tarjet1.remove(i)
tarjet1
```

```python
for i in tarjet1:
    if i == '':
        tarjet1.remove(i)
tarjet1
```

```python
mes=tarjet1[:75:3]
ano=tarjet1[1:75:3]
precio=tarjet1[2:75:3]

c= [re.sub(',', '.', i) for i in precio]
d=[float(i) for i in c]
```

```python
dic={
    'Precio(euros/mes)':d,}

luzPrecio=pd.DataFrame(dic)
luzPrecio
```
